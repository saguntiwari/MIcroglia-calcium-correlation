```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Load data from the Excel file
file_path = r'D:\PhD Manuscript\Figures\Soma Process Whole Cell\New Same ROI as old\Full Cell\3\pre_post_laser_data_with_spatial_info_in_um.xlsx'
post_laser_df = pd.read_excel(file_path, sheet_name='Post_Laser')

# folder path where the Excel file is located
folder_path = os.path.dirname(file_path)

## Laser injury coordinates (in pixels)(check the laser co-ordinates and and image orientation/size carefully,and put accrodingly)
laser_x = 250
laser_y = 221

# Convert to micrometers (0.497 um per pixel)
pixel_size = 0.497
laser_x_um = laser_x * pixel_size
laser_y_um = laser_y * pixel_size

# Calculate the distance of each event from the laser injury point
post_laser_df['Distance_from_Laser'] = np.sqrt((post_laser_df['Center_X_in_um'] - laser_x_um)**2 + 
                                                (post_laser_df['Center_Y_in_um'] - laser_y_um)**2)

# Adjust the x-axis to make 600 seconds/frame as 0 and increase accordingly
post_laser_df['Adjusted_Start_Time'] = post_laser_df['Start_Time'] - 600
post_laser_df['Adjusted_End_Time'] = post_laser_df['End_Time'] - 600  # If needed for plotting

# Calculate duration time (End Time - Start Time)
post_laser_df['Duration_Time'] = post_laser_df['Adjusted_End_Time'] - post_laser_df['Adjusted_Start_Time']

# Filter the data to include only the time between 600 and 610 seconds
post_laser_df_filtered = post_laser_df[(post_laser_df['Adjusted_Start_Time'] >= 0) & (post_laser_df['Adjusted_Start_Time'] <= 10)]

#figure
fig, ax = plt.subplots(figsize=(12, 8), dpi=1200)

# Plot each event, showing the start time and the duration
for i, row in post_laser_df_filtered.iterrows():
    # Plot a larger bright red dot at the end of the event duration
    ax.scatter(row['Adjusted_Start_Time'] + row['Duration_Time'], 
               row['Distance_from_Laser'], 
               color='red', alpha=1, s=450, edgecolors='black', linewidths=1)  # Larger red dots

# Set y-axis and x-axis limits to start from 0
ax.set_ylim(0, 140)  # Limit y-axis to 140
#ax.set_xlim(left=0)   # Ensure x-axis starts from 0
ax.set_xlim(0,400)   # Ensure x-axis starts from 0

# Adjust the layout to make sure everything fits
plt.tight_layout()

# Save the figure 
figure_path = os.path.join(folder_path, 'combined_figure_post_laser_with_end_vvvtimes_larger_dots_l.png')
plt.savefig(figure_path, dpi=200, transparent=True)  

# Show the combined figure
plt.show()
figure_path
```
